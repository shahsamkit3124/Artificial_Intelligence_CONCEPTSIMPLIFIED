### CHATBOT WITH RULE-BASED RESPONSES

Build a simple chatbot that responds to user inputs based on predefined rules. Use if-else statements or pattern matching
techniques to identify user queries and provide appropriate responses. This will give you a basic understanding of natural
language processing and conversation flow.

```python

def bot(user_ans):
    # Convert user input to lowercase
    user_ans = user_ans.lower()
    
    if "hiee" in user_ans or "hello" in user_ans or "hi" in user_ans:
        res = "Hello there! How can I assist you today?"
        
    elif "how's life going?" in user_ans or "how are you?" in user_ans:
        res = "I'm just a ChatBot, so I don't have personal experiences or emotions, but I'm here and ready to help you. Feel free to ask me anything!"
    
    elif "what do you study?" in user_ans or "what's your expertise?" in user_ans:
        res = "I specialize in processing and providing information. Feel free to ask me questions on a wide range of topics."
    
    elif "what's your name?" in user_ans:
        res = "I go by the name ChatBot."
    
    elif "what's my name?" in user_ans:
        res = "I'm sorry, but I don't have access to personal information, including your name."
    
    elif "tell me a joke" in user_ans or "say something funny" in user_ans:
        res = "Sure, here's a joke for you: Why don't scientists trust atoms? Because they make up everything!"
    
    elif "what's the weather like today?" in user_ans:
        res = "I'm afraid I can't provide real-time information like weather updates. Is there anything else you'd like to know?"
    
    elif "tell me an interesting fact" in user_ans:
        res = "Certainly! Did you know that honey never spoils? Archaeologists have found pots of honey in ancient Egyptian tombs that are over 3,000 years old and still perfectly edible."
    
    elif "can you recommend a book?" in user_ans:
        res = "Of course! How about 'The Hitchhiker's Guide to the Galaxy' by Douglas Adams? It's a classic science fiction comedy."
    
    elif "what's the meaning of life?" in user_ans:
        res = "Ah, the age-old question! According to 'The Hitchhiker's Guide to the Galaxy,' the answer is 42, though the actual question remains a mystery."
    
    elif "what's your favorite movie?" in user_ans:
        res = "I don't have personal preferences, but I've heard great things about 'The Matrix' and 'Inception'!"
    
    elif "who is your creator?" in user_ans or "who made you?" in user_ans:
        res = "I was created by Sanskriti Kadam."
    
    elif "what's the capital of France?" in user_ans:
        res = "The capital of France is Paris."
        
    elif "tell me about space exploration" in user_ans:
        res = "Space exploration is the investigation of outer space using astronomy, robotics, and human spaceflight. It has led to remarkable discoveries about our universe and beyond."
    
    elif "recommend a travel destination" in user_ans:
        res = "How about visiting Kyoto, Japan? It's known for its stunning temples, traditional tea houses, and beautiful cherry blossoms."
    
    elif "what the difference between a virus and a bacteria?" in user_ans:
        res = "Viruses are smaller and need a host to reproduce, while bacteria are single-celled organisms that can multiply on their own. Both play important roles in various ecosystems."
    
    elif "explain artificial intelligence" in user_ans:
        res = "Artificial Intelligence (AI) refers to the simulation of human intelligence in computers. It involves tasks like learning, problem-solving, and decision-making."
    
    else:
        res = "I'm sorry, I'm having trouble understanding your question. Could you please rephrase it or ask something else?"
    
    return res

```

```python
def Main():
    
    print("<---- Hello, Myself ChatBot !! ---->")
    print("Feel Free to Ask Anything !!")
    
    print("\n")
    
    while True:
        user_ans = input("Ask me here : ")
        
        if user_ans.lower() == 'bbyee':
            
            print("ChatBot : Bye, i hope you have a nice day! ðŸ˜Š")
            print("\n")
            
            break
            
        res = bot(user_ans)
        
        print("ChatBot:",res)
        print("\n")


if __name__ == "__main__":
    Main()
```
